```python
import pandas as pd
data=pd.read_csv('datasets/startup_funding.csv')
df=data.copy()
df
```

```python
amount=df.loc[(df['CityLocation']=='Bangalore') | (df['CityLocation']=='New Delhi'), 'AmountInUSD']
```

```python
amount.dropna(inplace=True)
```

```python
amount=amount.str.replace(',', '')
```

```python
amount=pd.to_numeric(amount)
```

```python
pop_mean=amount.mean()
```

```python
sample=amount.sample(50)
```

```python
sam_mean=sample.mean()
```

```python
sampling_error=pop_mean-sam_mean
sampling_error
```
